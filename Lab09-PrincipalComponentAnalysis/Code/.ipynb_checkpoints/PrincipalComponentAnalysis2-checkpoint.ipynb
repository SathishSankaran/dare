```python
import math

import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
### Configuration
dataFile = '../Data/FormationPrediction.csv'

# Data columns
colsInputs = ['GammaRay_API','BulkDensity_GPerCC','Resistivity_OHMSM','DensityPorosity_PCT','EffectivePorosity_PCT','TotalOrganicCarbon_WTPCT']
colsOutputs = ['Formation Class']
colsDrop = ['WellName','Latitude_BH','Longitude_BH','TopOfZone_FT','BottomOfZone_FT','Isopach_FT','TVD_FT']

# Split train and test data
testsize = 0.25
cv = 5
```

```python
### Data Loading (raw data)
df = pd.read_csv(dataFile)
df = df.drop(columns=colsDrop,axis='columns')
```

```python
### Data Pre-processing
inputs = df[colsInputs]
outputs = df[colsOutputs]

# Scaler
inputs = StandardScaler().fit_transform(inputs)

print(type(inputs))
```

```python
# Principal Component Analysis
clf = PCA(n_components=3)
model = clf.fit_transform(inputs)

clf.explained_variance_ratio_
```

```python
pcdf = pd.DataFrame(data=model,columns=['PC1','PC2','PC3'])
pcdf = pd.concat([pcdf,outputs[['Formation Class']]],axis=1)

fig = plt.figure(figsize=(40,10))
ax = fig.add_subplot(1,2,1)
for target,color in zip(['A','B'],['b','r']):
    indices = (pcdf['Formation Class'] == target)
    ax.scatter(pcdf[indices]['PC1'],pcdf[indices]['PC2'],c=color,s=50)
ax.legend(['A','B'])
ax.set_xlabel('PC1',fontsize=15)
ax.set_ylabel('PC2',fontsize=15)
ax.grid()

ax = fig.add_subplot(1,2,2,projection='3d')
for target,color in zip(['A','B'],['b','r']):
    indices = (pcdf['Formation Class'] == target)
    ax.scatter(pcdf[indices]['PC1'],pcdf[indices]['PC2'],pcdf[indices]['PC3'],c=color,s=50)
ax.legend(['A','B'])
ax.set_xlabel('PC1',fontsize=15)
ax.set_ylabel('PC2',fontsize=15)
ax.set_zlabel('PC3',fontsize=15)
ax.grid()
```
